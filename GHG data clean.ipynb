```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import geopandas as gpd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
ghg2013 = pd.read_csv('ghg2013.csv')
```

```python
ghg2018 = pd.read_csv('ghg2018.csv')
```

```python
print('ghg2013 has '+ str(len(ghg2013)) + " records")
print('ghg2018 has '+ str(len(ghg2018)) + " records")
```

```python
###  join by BIN
ghg2013_join = pd.DataFrame(columns=['BIN','GHG'])
ghg2013_join['BIN'] = ghg2013['BIN'].str.replace(',','')
ghg2013_join['GHG'] = ghg2013['Total GHG Emissions(MtCO2e)']
join_BIN = pd.merge(ghg2013_join, ghg2018, left_on="BIN", right_on="NYC Building Identification Number (BIN)")
#join_BIN
print("join by BIN has "+str(len(join_BIN[join_BIN[['GHG','Total GHG Emissions (Metric Tons CO2e)']].notnull().all(1)]))+" records")
```

```python
ghg2013.head(3)
```

```python
### join by BBL
ghg2013_join = pd.DataFrame(columns=['BBL','GHG'])
ghg2013_join['BBL'] = ghg2013['BBL'].astype(str).str[:-2]
ghg2013_join['GHG'] = ghg2013['Total GHG Emissions(MtCO2e)']
join_BBL = pd.merge(ghg2013_join, ghg2018, left_on="BBL", right_on="BBL - 10 digits")
print("join by BBL has "+str(len(join_BBL[join_BIN[['GHG','Total GHG Emissions (Metric Tons CO2e)']].notnull().all(1)]))+" records")
```

```python
ghg_merge = join_BBL[join_BIN[['GHG','Total GHG Emissions (Metric Tons CO2e)']].notnull().all(1)]
```

```python
ghg_merge.head(3)
```

```python
ghg_merge['Total GHG Emissions (Metric Tons CO2e)'] = pd.to_numeric(ghg_merge['Total GHG Emissions (Metric Tons CO2e)'],errors='coerce')
ghg_merge['GHG'] = pd.to_numeric(ghg_merge['GHG'],errors='coerce')
```

```python
ghg_merge['GHG change'] = (ghg_merge['Total GHG Emissions (Metric Tons CO2e)']-ghg_merge['GHG'])/ghg_merge['GHG']
```

```python
ghg_merge = ghg_merge[['BBL','GHG change','Borough','Self-Reported Gross Floor Area (ft²)',
                       'Largest Property Use Type','Year Built','ENERGY STAR Score','Occupancy','Census Tract']]
```

```python
ghg_merge.to_csv('ghg.csv')
```

```python
### merge with pluto data
pluto = pd.read_csv('pluto.csv')
pluto.head(3)
```

```python
pluto.columns
```

```python
pluto = pluto[['bbl','latitude', 'longitude','borough','ct2010','bldgclass','landuse','bldgarea','numfloors',
              'comarea','resarea', 'officearea', 'retailarea', 'garagearea', 'strgearea','factryarea',
              'assessland', 'assesstot','builtfar',]]
```
